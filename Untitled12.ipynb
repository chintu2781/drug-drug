```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
drugs = ['Aspirin', 'Ibuprofen', 'Warfarin', 'Paracetamol']
interactions = [
    ('Aspirin', 'Ibuprofen', 'Antagonistic'),
    ('Aspirin', 'Warfarin', 'Synergistic'),
    ('Ibuprofen', 'Paracetamol', 'Neutral'),
    ('Warfarin', 'Paracetamol', 'Antagonistic')
]
```

```python
for drug in drugs:
  G.add_node(drug)
for d1, d2, interaction in interactions:
  G.add_edge(d1, d2, interaction=interaction)
```

```python
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
```

```python
print("Degree Centrality:", degree_centrality)
print("Betweenness Centrality:", betweenness_centrality)
print("Closeness Centrality:", closeness_centrality)
```

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
```

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='lightblue')
nx.draw_networkx_edges(G, pos, width=2)
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')
edge_labels = nx.get_edge_attributes(G, 'interaction')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
plt.title("Drug Interaction Graph", fontsize=16)
plt.show()
```
